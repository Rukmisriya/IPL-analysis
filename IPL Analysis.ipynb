```python
# Converting data types of columns
df['second_ings_score'] = df['second_ings_score'].str.replace('"', '')
df['second_ings_score']=pd.to_numeric(df['second_ings_score'])
print(df.dtypes)
```

```python
# Renaming the columns
df.rename(columns={'first_ings_score':'first_innings_score','`player_of_the_match': 'match_of_the_player'},inplace=True)
df
```

```python
# taking new columns batted_first,batted_second
# lamda function is used to filter the teams
df['batted_first'] = df.apply(lambda row: row['toss_winner'] if row['toss_decision'] == 'Bat' else row['team2'] if row['toss_winner'] == row['team1'] else row['team1'], axis=1)
df['batted_second'] = df.apply(lambda row: row['toss_winner'] if row['toss_decision'] == 'Field' else row['team2'] if row['toss_winner'] == row['team1'] else row['team1'], axis=1)
df
```

```python
# team with the second-most number of first-innings scores
print(df[['batted_first', 'first_innings_score']].max())
print("--"*20)
#team with the second-most number of second-innings scores
print(df[['batted_second', 'second_ings_score']].max())
print("Rajastan is the team with most first-innings scores and second-innings scores")
```

```python
#sorting the first_ing wkts
df1 = df.sort_values(by='first_ings_wkts',ascending=True)
# printing last but one row
t1=df1[['batted_first', 'first_ings_wkts']].iloc[-2]
print(t1)
print("--"*20)
#sorting the second_ing wkts
df2 = df.sort_values(by='second_ings_wkts',ascending=True)
# printing last but one row
t2=df2[['batted_second', 'second_ings_wkts']].iloc[-2]
print(t2)
print(f"teams with the second-most number of second-innings wickets is {t2['batted_second']} and first-innings wickets is {t1['batted_first']}")
```

Insights

```python
import warnings
warnings.simplefilter(action="ignore" , category = FutureWarning)
# counting match winned
team_wins = df['match_winner'].value_counts()
team_wins.plot(kind='bar', color='skyblue')
# plotting no of wins vs teams
plt.title('Number of Matches Won by Each Team')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
```

```python
# Top Scorers
# grouping by top scorer and their high score (descending order)
top_scorers = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False)
top_scorers.plot(kind='bar', color='teal')
# plotting players vs their total scores
plt.title('Top Scorers in IPL Matches')
plt.xlabel('Players')
plt.ylabel('Total Scores')
plt.show()
```

```python
# Toss Decision Impact
# checking wins against toss decision ("bat" or "field")
toss_decision_wins = df[df['match_winner'] == df['toss_winner']]['toss_decision'].value_counts()
overall_toss_decision_counts = df['toss_decision'].value_counts()
print(overall_toss_decision_counts)
toss_decision_wins.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Number of Wins Based on Toss Decision')
plt.xlabel('Toss Decision')
plt.ylabel('Number of Wins')
```

Best Performing Teams: Teams with the most wins can be identified, giving insight into overall ipl.
